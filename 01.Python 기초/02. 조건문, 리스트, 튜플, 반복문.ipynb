### 조건문
- 상황에 따라 실행 흐름을 다르게 하는 문법
```python
if True : 
    코드블럭
elif True :
    코드블럭
else :
    코드블럭
```

- 콜론 = ":"과 들여쓰기 = "space*4", "Tab"가 매우 중요!
- 조건문 키워드 : if, elif, else

````python
score = int(input("점수 입력 >>"))
if score <= 100 and score > 90 :
    print(f"{score}점은 A학점 입니다.")
elif score <= 90 and score > 80 :
    print(f"{score}점은 B학점 입니다.")
elif score <= 80 and score > 70 :
    print(f"{score}점은 C학점 입니다.")    
elif score <= 70 and score > 60 :
    print(f"{score}점은 D학점 입니다.")
elif score < 60 :
    print(f"{score}점은 F학점 입니다.")
else :
    print("잘 못 입력하였습니다!")
````

````python
score = int(input("점수입력 >>")) #input() 결과값은 문자형이므로 형변환

if score <= 100 and score >= 0 :
    if score >= 90 :
        grade = "A" # A학점
    elif score >= 80 :
        grade = "B" # B학점
    elif score >= 70 :
        grade = "C" # C학점
    elif score >= 60 :
        grade = "D" # D학점
    else :
        grade = "F"
    print(f"{score}점은 {grade}학점 입니다.")
else : 
    print("잘못입력했어")
````

### 튜플(list)

### 반복문
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
```python
while 조건식 : # 반복횟수가 명확하지 않을 때
    코드블럭
    코드블럭

for 지역변수 in 시퀀스 자료구조 : # 반복횟수가 명확할 때
    코드블록
    코드블록
```
- break : 반복을 종료하고 , 아래로 빠져나오는 기능
- continue : 위 (조건식, 다음 요소)로 올라가는 기능

````python
user_input = True # 초기값 설정

while user_input : 
    user_input = bool(input("True, 공백입력 >> "))
````

````python
food_list = ["피자","햄버거","치킨"]
for food in food_list : 
    print(food)
````

````python
# range(시작값, 끝값,[, 보폭]) : 시작값부터 끝값까지 일련의 순서로 범위를 만들어내는 기능
# range([시작값, ] 끝값[, 보폭])

for i in range(0,5,2) :
    print(i)
````

````python
score_list = [90, 45 , 70, 60 ,55]
score_list
````

````python
for i in range(len(score_list)) : # range(len()) : 인덱스 번호로도 for문 활용이 가능!
    print(score_list[i])
````
