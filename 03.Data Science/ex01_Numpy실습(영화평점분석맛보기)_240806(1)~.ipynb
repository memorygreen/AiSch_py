```python
## 자영풀이

# 각 사용자별 평점 평균 구하기
# 1) 사용자 아이디의 고유값을 구해보자@
# 1-1) 유저 아이디 값이 담긴 열에 접근하자!
# 1-2) 고유값을 뽑아주는 함수를 사용해보자 > np.unique()
user_id = np.unique(data[:,0])
user_id
```

```python
## 자영풀이
#2) 처음부터 모든 아이디를 보기는 어렵다!
# > 유저 아이디가 1인 사람들의 데이터를 뽑아내보자 (불리언 인덱싱 사용)
user_1 = data[data[:,0] ==1]
# > 불리언 인덱싱으로 뽑힌 데이터 중에서 평점 데이터만 뽑아보자(슬라이싱)

# > 뽑은 평점 데이터에 평균을 구해보자.
user_1[:,2].mean()
```

```python
# 쌤풀이
id_ond_mean = data[data[:,0] ==1].mean()
```

```python
#3) 반복문을 이용해서 모든 사용자의 평점 평균을 구해보자
id_mean = []

for i in user_id:
    id_one = data[data[:,0] == i] # 사용자 아이디별로 데이터를 뽑아주는 단계
    id_one_mean = id_one[: , 2].mean() # 평점 데이터를 뽑아서 평균을 구해주는 단계
    id_mean.append([i , id_one_mean])
```
