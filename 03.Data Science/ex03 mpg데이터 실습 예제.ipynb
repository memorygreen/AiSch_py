```python
# Q1. 차량 카테고리별 도시 연비의 평균을 구해보자
# 차량 카테고리의 갯수부터 파악해보자
mpg['category'].value_counts()
# mpg.unique()
# mpg.groupby('category')
# mpg.loc[:,['cty','category']]
```

```python
# 필요한 컬럼들만 묶어주자 - groupby 함수
cty_mean = mpg[['cty','category']].groupby(by = 'category').mean()
cty_mean
```

```python
# 다른 방법으로 접근해보자 - agg함수 이용
# agg() : 다중집계 함수 / 여러 개의 값을 받아서 일괄적으로 연산을 진행하는 함수
mpg.groupby('category').agg(cty_mean = ('cty', 'mean'))
```

```python
cty_mean = mpg[['cty','category']].groupby('category').agg(cty_mean = ('cty', 'mean'))
cty_mean
```

```python
# Q2. 위에서 조사한 연비 평균 데이터를 이용해서 어떤 차종의 연비가 좋은지 내림차순으로 정렬
cty_mean.sort_values(by='cty_mean', ascending =False)
```

```python
# Q3. 어떤 회사의 자동차가 고속도로 연비가 가장 높을까? 상위 3개의 회사를 뽑아보자
hwy_mean = mpg[['manufacturer', 'hwy']].groupby(by = 'manufacturer').agg(hwy_mean = ('hwy', 'mean')).sort_values(by='hwy_mean', ascending = False)
hwy_mean
```

```python
# 상위 3개 회사 뽑기
hwy_mean.head(3)
hwy_mean.iloc[:3]
```

```python
# compact 차량을 가장 많이 생산하는 회사별로 데이터를 나열해라
# 1. category가 compact인 차종만 뽑아오자
compact = mpg[mpg['category'] =='compact' ]
compact
# 2. 회사별로 묶은 후 집계 (숫자 세기)
result = compact.groupby(by='manufacturer').agg(cmp_cnt = ('category', 'count'))
result
# 3. 정렬
result.sort_values(by = 'cmp_cnt', ascending = False)
```

```python
# query 함수 사용하기
# 차종정보 compact인 차량 접근 > 제조사별 그룹핑
mpg.query('category == "compact"')['manufacturer'].value_counts()
mpg.query('category == "compact"').value_counts('manufacturer')
```
