```python
rs = hr_info[hr_info['EDU_LEVEL'] == '박사 학위']
rs['SALARY'].mean()
```

```python
# 모든 학력을 돌아보면서 평균 연봉을 구해봅시다
for edu_level in hr_info['EDU_LEVEL'].unique() :
    rs = hr_info[hr_info['EDU_LEVEL'] == edu_level]
    print(f"{edu_level} : {rs['SALARY'].mean()}")
```

```python
# 데이터 그룹화를 통한 통계 분석
# groupby 함수를 이용하는 실습
hr_info[['SALARY', 'EDU_LEVEL']].groupby('EDU_LEVEL')
```

```python
for idx, df in hr_info[['SALARY', 'EDU_LEVEL']].groupby('EDU_LEVEL') :
    display(idx)
    display(df)
```

```python
# 두개 이상의 기준으로 그룹화
# 교육 수준과 성별에 따른 평균 연봉 확인
hr_info[['SALARY', 'EDU_LEVEL', 'GENDER']].groupby(['EDU_LEVEL' ,'GENDER']).mean()
```

```python
# 다중 집계 함수 agg 사용해보기
df_g = hr_info[['SALARY', 'EDU_LEVEL', 'GENDER']].groupby(['EDU_LEVEL' ,'GENDER'])
df_g.agg(['mean', 'max', 'min', 'median'])
```

```python
# 국적에 따른 평균 성과점수와 최대 연봉 확인해보기
hr_info.groupby('NATION').agg({
    'PERFORMANCE_SCORE' : 'mean',
    'SALARY' : 'max'
})
```

### 상관계수 분석 (상관관계 확인)
- 두 변수 사이의 관계를 수치화 시켜서 살펴보는 방법


### 분산
- 변수가 기댓값으로부터 얼마나 떨어진 곳에 분포하는지 가늠하는 숫자
- 데이터가 얼마나 뭉쳐있는지 / 흩어져 있는지 숫자로 확인하는 하나의 지표

- 공분산 값이 0보다 크다 > 두 수는 비례하는 관계
- 공분산 값이 0보다 작다 > 두 수는 반비례하는 관계
- 공분산 값이 0에 가깝다 > 서로 영향을 주지 않는다.

#### 상관계수
- 공분산 값을 -1 ~ 1사이로 정규화 한 값이다.

- 상관계수 해석 가이드(일반론)
  - 우선 절대값을 사용한다(중요한 파트)
  - 0 <= 절대값 < 0.2 : 상관관계가 없다.
  - 0.2 <= 절대값 < 0.4 : 약한 상관관계가 있다.
  - 0.4 <= 절대값 < 0.6 : 보통의 상관관계가 있다.
  - 0.6 <= 절대값 < 0.8 : 강한 상관관계가 있다.
  - 0.8 <= 절대값 < 1 : 매우 강한 상관관계가 있다.

```python
# 직원 성과점수와 연봉에 상관계수를 확인
# .corr() : 상관계수를 확인하는 함수
hr_info[['PERFORMANCE_SCORE', 'SALARY']].corr()
```
