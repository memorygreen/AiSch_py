```python
#2.  
# data2019, data2020, data2021
#  '' ->  
# encoding= 'euc-kr'


data2019 = pd.read_csv('data/2019.csv',  encoding='euc-kr')
data2020 = pd.read_csv('data/2020.csv',  encoding='euc-kr')
data2021 = pd.read_csv('data/2021.csv',  encoding='euc-kr')

```

```python
 # 3.       (2021 ->)

data2021 = data2021[data2021['']!='']
data2021
```

```python
#4.        ''  

data2021[''] = data2021.loc[:,'':''].sum(axis=1)
```

```python
data2020[''] = data2020.loc[:,'':''].sum(axis=1)
data2020
```

```python
data2019[''] = data2019.loc[:,'':''].sum(axis=1)
data2019
```

```python
#5.   '   ''     
data2019_cnt = data2019[data2019['']=='']
data2019_cnt

data2019_cnt.set_index( pd.Series(['','','','','','']), inplace=True)
data2019_cnt
```

```python
data2020_cnt = data2020[data2020['']=='']
data2020_cnt

data2020_cnt.set_index( pd.Series(['','','','','','']), inplace=True)
data2020_cnt
```

```python
data2021_cnt = data2021[data2021['']=='']
data2021_cnt

data2021_cnt.set_index( pd.Series(['','','','','','']), inplace=True)
data2021_cnt
```

```python
#6.      -> Series   .
data2021_sum  = data2021_cnt['']
data2021_sum
```

```python
data2020_sum  = data2020_cnt['']
data2020_sum
```

```python
data2019_sum  = data2019_cnt['']
data2019_sum
```

```python
#7.    (2019, 2020, 2021)
df = pd.DataFrame()
df['2019'] = data2019_sum
df['2019'].index= ['','','','','','']
# df['2020'] = data2020_sum
# df['2021'] = data2021_sum
df['2019']
```

```python
data2020_rate = (data2020_sum - data2019_sum) / data2019_sum * 100
df['2019-2020 '] = data2020_rate


df['2020'] = data2020_sum
df
```

```python

data2021_rate = (data2021_sum - data2020_sum) / data2020_sum * 100
df['2020-2021 '] = data2021_rate

df['2021'] = data2021_sum
df

#  -  /  * 100
```
