- 범죄데이터를 활용하여 각 연도별 범죄 증감률을 구해보자


```python
# 1. 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
#2. 데이터 불러오기
#변수명 data2019, data2020, data2021
#컬럼 중 '관서명' -> 인덱스 설정
#인코딩 encoding= 'euc-kr'


data2019 = pd.read_csv('data/2019.csv',  encoding='euc-kr')
data2020 = pd.read_csv('data/2020.csv',  encoding='euc-kr')
data2021 = pd.read_csv('data/2021.csv',  encoding='euc-kr')

```

```python
data2019
```

```python
data2020
```

```python
data2021
```

```python
 # 3. 한 개의 년도에만 존재하는 경찰청 삭제 (2021년도 ->광주지방경찰청)

data2021 = data2021[data2021['관서명']!='광주지방경찰청']
data2021
```

```python
#4. 각 연도별로 범죄 발생 수의 합계를 구하여 '총합' 컬럼을 추가하기

data2021['총합'] = data2021.loc[:,'살인':'폭력'].sum(axis=1)
```

```python
data2021
```

```python
data2020['총합'] = data2020.loc[:,'살인':'폭력'].sum(axis=1)
data2020
```

```python
data2019['총합'] = data2019.loc[:,'살인':'폭력'].sum(axis=1)
data2019
```

```python
#5. 각 연도별로 구분' 컬럼의 값이 '발생건수' 인 데이터만 출력하여 변수에 담기
data2019_cnt = data2019[data2019['구분']=='발생건수']
data2019_cnt

data2019_cnt.set_index( pd.Series(['광주지방경찰청계','광주동부경찰서','광주서부경찰서','광주남부경찰서','광주북부경찰서','광주광산경찰서']), inplace=True)
data2019_cnt
```

```python
data2020_cnt = data2020[data2020['구분']=='발생건수']
data2020_cnt

data2020_cnt.set_index( pd.Series(['광주지방경찰청계','광주동부경찰서','광주서부경찰서','광주남부경찰서','광주북부경찰서','광주광산경찰서']), inplace=True)
data2020_cnt
```

```python
data2021_cnt = data2021[data2021['구분']=='발생건수']
data2021_cnt

data2021_cnt.set_index( pd.Series(['광주지방경찰청계','광주동부경찰서','광주서부경찰서','광주남부경찰서','광주북부경찰서','광주광산경찰서']), inplace=True)
data2021_cnt
```

```python
#6. 각 연도별로 총합 데이터만 추출 -> Series 형태로 출력될 것입니다.
data2021_sum  = data2021_cnt['총합']
data2021_sum
```

```python
data2020_sum  = data2020_cnt['총합']
data2020_sum
```

```python
data2019_sum  = data2019_cnt['총합']
data2019_sum
```

```python
#7. 총합 데이터 이름 변경해주기(2019총계, 2020총계, 2021총계)
df = pd.DataFrame()
df['2019총계'] = data2019_sum
df['2019총계'].index= ['광주지방경찰청계','광주동부경찰서','광주서부경찰서','광주남부경찰서','광주북부경찰서','광주광산경찰서']
# df['2020총계'] = data2020_sum
# df['2021총계'] = data2021_sum
df['2019총계']
```

```python
data2020_rate = (data2020_sum - data2019_sum) / data2019_sum * 100
df['2019-2020 증감율'] = data2020_rate


df['2020총계'] = data2020_sum
df
```

```python

data2021_rate = (data2021_sum - data2020_sum) / data2020_sum * 100
df['2020-2021 증감율'] = data2021_rate

df['2021총계'] = data2021_sum
df

# 금년 - 작년 / 작년 * 100
```
