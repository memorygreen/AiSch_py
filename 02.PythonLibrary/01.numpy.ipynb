### BooleanIndexing(ë¶ˆë¦¬ì–¸ì¸ë±ì‹±)
- ì¡°ê±´ì— í•´ë‹¹ë˜ëŠ”(True)ë°ì´í„°ë§Œ ì¶”ì¶œí•  ë•Œ ì‚¬ìš©í•˜ëŠ” ë°©ë²•

```python
# ì„ì˜ì˜ ì •ìˆ˜ ë°ì´í„°ë¥¼ ìƒì„±
score = np.array([80,75,55,96,30])
score # array([80, 75, 55, 96, 30])
 
```

```python
# ì„¸ë²ˆì§¸ì‹œê°„
```

```python
# ì ìˆ˜ê°€ 80ì  ì´ìƒì¸ ë°ì´í„°ë§Œ ì¶”ì¶œ!
score >=80 # array([ True, False, False,  True, False])
score[score >= 80] # array([80, 96])
# ë³€ìˆ˜ëª…[ì¡°ê±´] -> ì¡°ê±´ì´ ì°¸ì¸ ë°ì´í„°ë§Œ ì¶”ì¶œ
# ê° ì ìˆ˜ ë°ì´í„°ì— ëŒ€ì‘ ì´ë¦„ ë°°ì—´ ìƒì„±
name = np.array(['í•˜ì€', 'í˜„ë¹ˆ', 'ìœ ë¯¼', 'ì±„ì›','ì§€í˜„'])
name #array(['í•˜ì€', 'í˜„ë¹ˆ', 'ìœ ë¯¼', 'ì±„ì›', 'ì§€í˜„'], dtype='<U2')

```

```python
# ì ìˆ˜ê°€ 80ì  ì´ìƒì¸ ì¹œêµ¬ì˜ ì´ë¦„ ì¶”ì¶œ
name[score>=80] # array(['í•˜ì€', 'ì±„ì›'], dtype='<U2')
```

```python
# ì ìˆ˜ê°€ 80ì  ì´ìƒì¸ ì¹œêµ¬ì˜ ì´ë¦„ ì¶”ì¶œ
# ì¶œë ¥í•˜ê³  ì‹¶ì€ ë°ì´í„°ê°€ ë“¤ì–´ìˆëŠ” ë°°ì—´ëª…[ì¡°ê±´]
# ë°°ì—´ëª…[ True ê°’ë§Œ ë½‘ì•„ì˜¬ ìˆ˜ ìˆëŠ” ì¡°ê±´ì‹] # 240805 ë°ì´í„°ì‚¬ì´ì–¸ìŠ¤(1-2) ì¶”ê°€
name[score >= 80] # array(['í•˜ì€', 'ì±„ì›'], dtype='<U2')
```

### ì‚¬ìš©ìë³„ ì˜í™” í‰ì  í‰ê·  êµ¬í•˜ê¸°

```python
# 1ë²ˆ ì‚¬ìš©ìë§Œ ì¶”ì¶œ - ë¶ˆë¦¬ì–¸ ì¸ë±ì‹±
# ì¡°ê±´ : ì‚¬ìš©ìì•„ì´ë””ê°€ 1ë²ˆì´ë‹ˆ?
data[data[:,0]==1][:,2].mean() # 4.188679245283019
```

### ì „ì²´ ì‚¬ìš©ìë³„ ì˜í™” í‰ì  í‰ê· 

```python
# ë°˜ë³µë¬¸ ì§„í–‰ì„ ìœ„í•´ì„œ ì‚¬ìš©ìì˜ ìˆ˜ í™•ì¸ -> ì¤‘ë³µëœ ë°ì´í„°ë¥¼ ì œê±°í•˜ê³  ì¶œë ¥í•˜ëŠ” ë°©ë²•
user_id = data[:,0] # array([   1,    1,    1, ..., 6040, 6040, 6040])
user_id_uni = np.unique(user_id) # array([   1,    2,    3, ..., 6038, 6039, 6040])
# ë°ì´í„°ì˜ ê¸¸ì´ í™•ì¸
len(user_id_uni) # 6040
```

```python
# ì:ë°˜ë³µë¬¸ ì²˜ë¦¬ (1ë²ˆë§Œ ë‹¤ë¥¸ ë²ˆí˜¸ë¡œ ë°”ë€Œë©´ ë¨)

for i in range(1, len(user_id_uni)):
    user = data[data[:,0]==i]
    print(f'{i}ë²ˆ ì‚¬ìš©ì í‰ì  : {user[:,2].mean()}')
```
